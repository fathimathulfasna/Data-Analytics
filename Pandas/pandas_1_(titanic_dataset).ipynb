#**Dictionry as columns and rows using pandas**

```python
data1 = pd.DataFrame(
    {
        'key1' : [1,2,4,5,6],
        'key2' : [4,5,6,7,8],
        'key3' : [3,4,5,6,6]
    }
)
```

```python
data1
```

```python
data2 = pd.DataFrame(
    {
        'key1' : [1,2,45,6,67],
        'key4' : [56,5,6,7,8],
        'key5' : [3,56,5,6,6]
    }
)
```

```python
data2
```

#**merging two dictionry dataframes**

```python
pd.merge(data1,data2)
```

```python
pd.merge(data1,data2, how ='left')
```

```python
pd.merge(data1,data2, how ='right')
```

#**Cross frames of two different Dictionary Dataframes**

```python
pd.merge(data1,data2, how ='cross')
```

#**concatination**

```python
pd.concat([data1,data2],axis=1)
```

#**indexing in dictionary data frames**

```python
data1 = pd.DataFrame({'key1': [1,2,4,5,6],
                       'key2' :[4,5,6,7,8],
                      'key3' :[3,4,5,6,6]},
                     index = ['a','b','c','d','e']
                     )
data1
```
